```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de polarización
def P_tau_minus(eta, beta_R):
    tanh_eta = np.tanh(eta)
    numerator = ((tanh_eta + 1)**2) * (beta_R**4 - 1)
    denominator = ((tanh_eta + 1)**2) * (1 + beta_R**4) + 8 * beta_R**2
    return numerator / denominator

# Rango de eta
eta_vals = np.linspace(-2.5, 2.5, 500)

# Valores de beta_R a graficar
beta_R_vals = [0.1, 0.3, 0.5, 0.7, 0.9]

# Crear la figura
width =3.6
plt.figure(figsize=(4* width/3, width))

# Graficar para cada beta_R
for beta_R in beta_R_vals:
    P_vals = P_tau_minus(eta_vals, beta_R)
    plt.plot(eta_vals, P_vals, label=fr'$\beta_R = {beta_R}$')

# Estética del gráfico
plt.xlabel(r'$\eta$', loc='right')
plt.xlim(-2.5, 2.5)
plt.ylabel(r'$\mathcal{P}_{\tau^-}$', loc='top', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.legend()
plt.grid()
plt.grid(which="minor", color="0.9")
plt.tight_layout()

plt.savefig('P_vlQ_tau_minus_vs_eta.pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de polarización
def P_tau_minus_avg(eta, beta_R):
    return (P_tau_minus(eta, beta_R) + P_tau_minus(-eta, beta_R)) / 2.0

# Rango de eta
eta_vals = np.linspace(-2.5, 2.5, 500)

# Valores de beta_R a graficar
beta_R_vals = [0.01, 0.1, 0.3, 0.5, 0.7, 0.9]

# Crear la figura
width =3.6
plt.figure(figsize=(4* width/3, width))

# Graficar para cada beta_R
for beta_R in beta_R_vals:
    P_vals = P_tau_minus_avg(eta_vals, beta_R)
    plt.plot(eta_vals, P_vals, label=fr'$\beta_R = {beta_R}$')

# Estética del gráfico
plt.xlabel(r'$\eta$', loc='right')
plt.xlim(-2.5, 2.5)
plt.ylabel(r'$\mathcal{P}_{\tau^-}$', loc='top', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.legend()
plt.grid()
plt.grid(which="minor", color="0.9")
plt.tight_layout()

plt.savefig('P_vlQ_tau_minus_vs_eta_avg.pdf')
plt.show()
```
